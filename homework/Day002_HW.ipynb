### 根據範例檔案的結果：

1. 取出班次一的每一個時間
2. 將班次一的每一個時間用一種資料型態保存
3. 將班次一到五與其所有時間用一種資料型態個別保存

```python
import csv

ROW = []

# 開啟 CSV 檔案
with open('./example.csv', newline='') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)
    
    # 以迴圈輸出每一列
    for i in rows:
        ROW.append(i)

ROW
```

```python
# 1. 取出班次一的每一個時間



for i in range(0, len(ROW[1])):
    if ':' in ROW[1][i]:
        print(ROW[1][i])

```

```python
# 2. 將班次一的每一個時間用一種資料型態保存

bus1_time = []

for i in range(0, len(ROW[1])):
    if ':' in ROW[1][i]:
        bus1_time.append(ROW[1][i])

bus1_time
```

```python
# 3. 將班次一到五與其所有時間用一種資料型態個別保存

bus1_time = []
bus2_time = []
bus3_time = []
bus4_time = []
bus5_time = []

for i in range(0, len(ROW[1])):
    if ':' in ROW[1][i]:
        bus1_time.append(ROW[1][i])
        
for i in range(0, len(ROW[2])):
    if ':' in ROW[2][i]:
        bus2_time.append(ROW[2][i])
        
for i in range(0, len(ROW[3])):
    if ':' in ROW[3][i]:
        bus3_time.append(ROW[3][i]) 
        
for i in range(0, len(ROW[4])):
    if ':' in ROW[4][i]:
        bus4_time.append(ROW[4][i])
        
for i in range(0, len(ROW[5])):
    if ':' in ROW[5][i]:
        bus5_time.append(ROW[5][i])

```
